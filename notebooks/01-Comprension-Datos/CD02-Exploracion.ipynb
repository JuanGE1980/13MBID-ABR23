##### Lectura de los datasets

```python
df_tarjetas = pd.read_csv("../../data/raw/datos_tarjetas.csv", sep=";")
df_tarjetas.head(1)
```

----

##### Descripción de los datasets

```python
def reporte_descripcion_dataset(df):
    columnas = df.columns
    print("Columnas del dataset:\n")
    for col in columnas:
        print(col)
    print(f"\nCantidad de filas: {df.shape[0]}")
```

```python
print("Descripción del dataset 'datos_creditos'")
reporte_descripcion_dataset(df_creditos)
```

```python
print("Descripción del dataset 'datos_tarjetas'")
reporte_descripcion_dataset(df_tarjetas)
```

```python
# Snippet obtenido desde d-tale para datos_creditos

# DISCLAIMER: 'df' refers to the data you passed in when calling 'dtale.show'

import numpy as np
import plotly.graph_objs as go

if isinstance(df_creditos, (pd.DatetimeIndex, pd.MultiIndex)):
	df = df_creditos.to_frame(index=False)

# remove any pre-existing indices for ease of use in the D-Tale code, but this is not required
df = df_creditos.reset_index().drop('index', axis=1, errors='ignore')
df.columns = [str(c) for c in df.columns]  # update columns to strings in case they are numbers

s = df['edad'].fillna('Missing').astype(str)
result = pd.value_counts(s).to_frame(name='data')
result['percent'] = (result['data'] / result['data'].sum()) * 100
result.index.name = 'edad'
result = result.rename(columns={'data': 'Frequency', 'percent': 'Percent'})
result = result[result['Frequency'] > 0]
result = result.reset_index().sort_values(['Frequency', 'edad'], ascending=[False, True])
#result = result.head(self.top)

# If you're having trouble viewing your chart in your notebook try passing your 'chart' into this snippet:
#
# from plotly.offline import iplot, init_notebook_mode
#
# init_notebook_mode(connected=True)
# chart.pop('id', None) # for some reason iplot does not like 'id'
# iplot(chart)
```
